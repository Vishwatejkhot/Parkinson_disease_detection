```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import *
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Parkinsson disease.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Check missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
ax = sns.countplot(x="status", data=df)
print(df.status.value_counts())
print(df.shape)
```

```python
minmax = MinMaxScaler()
X = df.drop(['status','name'],axis='columns')
y = df["status"]
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train = minmax.fit_transform(X_train)
X_test = minmax.transform(X_test)
```

# Logistic Regression

```python

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=16)
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test)
```

```python
y_pred = logreg.predict(X_test)
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10,7))
sns.heatmap(cnf_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python

print('Logistic regression accuracy:', accuracy_score(y_test, y_pred ))
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

DTC=DecisionTreeClassifier()
DTC.fit(X_train, y_train)
y_pred_DTC = DTC.predict(X_test)

```

```python
cnf_dtc_matrix = metrics.confusion_matrix(y_test, y_pred_DTC)

plt.figure(figsize=(10,7))
sns.heatmap(cnf_dtc_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

```

```python
# Get feature importances from the Decision Tree model
feature_importances = DTC.feature_importances_

# Create a DataFrame to store feature names and their importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the features by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Create a bar chart to visualize feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title('Feature Importances from Decision Tree Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python

print('Decision tree accuracy:',accuracy_score(y_test, y_pred_DTC))


```
